```python
import pandas as pd
from google.cloud import storage
import urllib.request

def download_csv_to_gcp_bucket(url, filename):
    """Downloads a CSV file from a Google Drive link and writes it to a GCP bucket."""

    # Set the name of the GCP bucket where you want to store the file
    bucket_name = 'globant-challenge'

    # Download the CSV file from the Google Drive link
    response = urllib.request.urlopen(url)
    df = pd.read_csv(response)

    # Create a GCP storage client and get the specified bucket
    storage_client = storage.Client(project='arboreal-stage-385814')
    bucket = storage_client.get_bucket(bucket_name)

    # Write the CSV file to the bucket
    blob = bucket.blob(filename)
    blob.upload_from_string(df.to_csv(index=False), content_type='text/csv')
    print('File {} uploaded to {}.'.format(filename, bucket_name))
```

```python
# Set the URL of the Google Drive CSV file
url = 'https://drive.google.com/file/d/1CHWAh4KQ3Mq9AJdY6a9mTSIn7HjYUtEY/view'

# Set the name of the file in the GCP bucket
filename = 'my_file.csv'

download_csv_to_gcp_bucket(url,filename)
```
